```python
import pandas as pd 
import numpy as np
data = {'numbers': [1,2,3,4,5,np.nan,6,7,np.nan,np.nan,8,9,10,np.nan]}
df = pd.DataFrame(data,columns=['numbers'])
df
```

```python
#replacing nan with median value
df.fillna(df.median())
```

```python
#replacing nan with mode value
data = {'numbers': [1,2,3,4,5,np.nan,6,7,np.nan,np.nan,8,9,10]}
df1 = pd.DataFrame(data,columns=['numbers'])

#mode
df1.fillna(df1.mode())
```

Q)What is the difference between merging and joining in dataframe?

    Ans)We can use join and merge to combine 2 dataframes. The join method works best when we are joining dataframes on their indexes. The merge method is more versatile and allows us to specify columns besides the index to join on for both dataframes.    
  

Q) Create a data frame and add a column. Add names in it and use apply function to return names whose length is greater than 5

```python
d = ['Jimin', 'Jin', 'Taehyung', 'Jungkook', 'Namjoon']
df = pd.DataFrame(data=d, columns=['names'])
df[df['names'].apply(lambda x: len(x)>5)]
```

Q) Create a numpy array and a dataframe and concatenate them.

```python
arr = np.array((1, 2, 3, 4, 5))
df = pd.DataFrame([6, 7, 8, 9, 10], columns=list('A'))
df = pd.concat([df, pd.DataFrame(arr, columns = ['B'])], axis=1)
df
```
